```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi
import zipfile

TITANIC_PATH = r"D:/Sumon/Coding/Datasets/Titanic"
ZIP_FILE = os.path.join(TITANIC_PATH, "titanic.zip")

def fetch_titanic_data():
    api = KaggleApi()
    api.authenticate()

    os.makedirs(TITANIC_PATH, exist_ok=True)
    print(f"Downloading dataset.....")
    api.competition_download_files("titanic", TITANIC_PATH)
    print(f"Data downloaded and extracted to {TITANIC_PATH}")

    with zipfile.ZipFile(ZIP_FILE, "r") as zip_ref:
        zip_ref.extractall(TITANIC_PATH)
    print(f"Data extracted to {TITANIC_PATH}")
fetch_titanic_data()
```

```python
import pandas as pd

def load_titanic_data(titanic_path=TITANIC_PATH):
    train_path = os.path.join(titanic_path, "train.csv")
    test_path = os.path.join(titanic_path,"test.csv")
    return pd.read_csv(train_path), pd.read_csv(test_path)
```

```python
train, test = load_titanic_data()
```

```python
train.head()
```

```python
train.info()
```

```python
test.info()
```

```python
train["Title"]=train["Name"].str.extract(r'([A-Za-z]+)\.',expand=False)
test["Title"]=test["Name"].str.extract(r'([A-Za-z]+)\.',expand=False)
```

```python
train["Title"].value_counts()
```

```python
train.drop("Name", axis=1, inplace=True)
test.drop("Name", axis=1, inplace=True)
```

```python
train.head()
```

```python
train.set_index("PassengerId")
test.set_index("PassengerId")
```

```python
train["cabin_number"]=train["Cabin"].str.extract(r'(\d+)', expand=False)
train["cabin_number"] = pd.to_numeric(train["cabin_number"], errors='coerce', downcast='integer')
train["Deck"]=train["Cabin"].str.extract(r'([A-Za-z])', expand=False)
train["Deck"].fillna('Unknown', inplace=True)
test["cabin_number"]=test["Cabin"].str.extract(r'(\d+)', expand=False)
test["cabin_number"] = pd.to_numeric(test["cabin_number"], errors='coerce', downcast='integer')
test["Deck"]=test["Cabin"].str.extract(r'([A-Za-z])', expand=False)
test["Deck"].fillna('Unknown', inplace=True)
```

```python
train["Deck_known"] = train["Deck"].apply(lambda x: 1 if x != 'Unknown' else 0)
test["Deck_known"] = test["Deck"].apply(lambda x: 1 if x != 'Unknown' else 0)
```

```python
train.head()
```

```python
train["ticket_count"]=train.groupby("Ticket")["Ticket"].transform("count")
test["ticket_count"]=test.groupby("Ticket")["Ticket"].transform("count")
```

```python
title_mapping = {
    'Mr':'Mr',
    'Mrs':'Mrs',
    'Miss':'Miss',
    'Master':'Master',
    'Dr':'Dr',
    'Col':'Officer',
    'Major':'Royalty',
    'Lady':'Royalty',
    'Sir':'Royalty',
    'Don':'Royalty',
    'Mme':'Mrs',
    'Ms':'Mrs',
    'COuntess':'Royalty',
}
train["Title"]=train["Title"].map(title_mapping)
test["Title"]=test["Title"].map(title_mapping)
```

```python
train['Fare_per_person'] = train["Fare"] / train["ticket_count"]
test['Fare_per_person'] = test["Fare"] / test["ticket_count"]
```

```python
train["Family_size"]=train["SibSp"]+train["Parch"]
test["Family_size"]=test["SibSp"]+test["Parch"]
```

```python
train["Is_Alone"]=train["Family_size"].apply(lambda x: 1 if x==0 else 0)
test["Is_Alone"]=test["Family_size"].apply(lambda x: 1 if x==0 else 0)
```

```python
import pandas as pd

bins = [0, 12, 18, 60, float('inf')]
labels = ['Child', 'Teen', 'Adult', 'Senior']
train["Age_Binned"] = pd.cut(train["Age"], bins=bins, labels=labels, right=False)
train["Age_Binned"] = train["Age_Binned"].cat.add_categories('Unknown').fillna('Unknown')
test["Age_Binned"] = pd.cut(test["Age"], bins=bins, labels=labels, right=False)
test["Age_Binned"] = test["Age_Binned"].cat.add_categories('Unknown').fillna('Unknown')
```

```python
train.head()
```

```python
train.drop(["Ticket","Cabin","SibSp","Parch"],axis=1,inplace=True)
test.drop(["Ticket","Cabin","SibSp","Parch"],axis=1,inplace=True)
```

```python
train.drop("Age",axis=1, inplace=True)
test.drop("Age",axis=1, inplace=True)
```

```python
train.head()
```

```python
X=train.drop(columns="Survived")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler

numerical_features = ["Pclass","Fare","cabin_number","Family_size","ticket_count","Deck_known","Is_Alone","Fare_per_person"]
categorical_features = ["Deck","Sex","Embarked","Age_Binned","Title"]

numerical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy='mean')),
    ("scaler", StandardScaler()),
])

categorical_transformer = Pipeline(steps=[
    ("sim_imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore")),
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)
```

```python
X_train = preprocessor.fit_transform(X)
X_train
```

```python
X_test = preprocessor.transform(test)
y_train = train["Survived"]
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
forest_clf.fit(X_train, y_train)
```

```python
y_pred = forest_clf.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
forest_score = cross_val_score(forest_clf, X_train, y_train, cv=10)
forest_score
```

```python
from sklearn.svm import SVC

svc_clf = SVC()
svc_clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score

svc_score = cross_val_score(svc_clf, X_train, y_train, cv=10)
svc_score
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

param_grid_rf = {
    'classifier__n_estimators': [100, 200, 500],
    'classifier__max_depth': [10, 20, None],
    'classifier__min_samples_split': [2, 5],
    'classifier__class_weight': ['balanced', None],

}

svc_param_grid = {
    'classifier__C': [0.1, 1, 10, 100],
    'classifier__kernel': ['linear', 'rbf', 'poly'],
    'classifier__gamma': ['scale', 'auto'],
    'classifier__degree': [3, 4, 5],  # For polynomial kernel
    'classifier__coef0': [0, 0.5], 
}

pipeline_rf = Pipeline([
    ('classifier', RandomForestClassifier(random_state=42)),
])

pipeline_svc = Pipeline([
    ('classifier', SVC(random_state=42)),
])

grid_rf = GridSearchCV(estimator=pipeline_rf, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2,scoring='accuracy')
grid_svc = GridSearchCV(estimator=pipeline_svc, param_grid=svc_param_grid, cv=5, n_jobs=-1, verbose=2,scoring='accuracy')
```

```python
grid_rf.fit(X_train, y_train)
```

```python
grid_svc.fit(X_train,y_train)
```

```python
print(f"Best parameters for Random Forest: {grid_rf.best_params_}")
print(f"Best cross-validation score for Random Forest: {grid_rf.best_score_}")
print(f"Best parameters for SVC: {grid_svc.best_params_}")
print(f"Best cross-validation score for SVC: {grid_svc.best_score_}")
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

estimators = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('svc', SVC(C=1, kernel='rbf', random_state=42)),
]

stacking_clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
```

```python
stacking_clf.fit(X_train, y_train)
```

```python
cv_scores = cross_val_score(stacking_clf, X_train, y_train, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean cross-validation score: {cv_scores.mean()}")
```
